```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('kc_house_data1.csv')
df
```

```python
df.dtypes
```

```python
df.describe()
```

Датасет показывает основную тенденцию ценообразования дома с точки зрения его местоположения, района постройки, интерьера и т. д. Есть индекс записи, id записи, дата,цена, кол-во спален, жилая площадь и т.д.

- Изучить стоимость недвижимости

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['price'])
plt.title('Стоимость недвижимости')
plt.xlabel('Стоимость, у.е.')
plt.ylabel('Количество домов, шт.')
```

- Изучить распределение жилой квадратуры

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['sqft_living'])
plt.title('Жилая площадь, кв.м')
plt.xlabel('Площадь, кв.м.')
plt.ylabel('Количество домов, шт.')
```

- Изучить рапределение года постройки

```python
counts = df['yr_built'].value_counts()
plt.scatter(x=counts.index, y=counts.values)
plt.title('Распределение застройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество домов, шт.')
```
